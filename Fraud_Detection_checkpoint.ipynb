```python
from scipy.stats import zscore

# Calculate z-scores for numeric features
numeric_features = data.select_dtypes(include=['float64', 'int64']).columns
z_scores = data[numeric_features].apply(zscore)

# Identify outliers (z-score > 3 or < -3)
outliers = (z_scores.abs() > 3).sum()
print("Outliers detected per feature:")
print(outliers)
```

```python
import scipy.stats as stats
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(outliers, kde=True)
plt.title('Original Data - Histogram with KDE')

plt.subplot(1, 2, 2)
stats.probplot(outliers, dist="norm", plot=plt)
plt.title('Original Data - Q-Q plot')

plt.tight_layout()
plt.savefig("outlier.png", dpi=80, bbox_inches='tight')
plt.show()
```

```python
# Import all theses Models
# Basic Classifiers
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB

# XGBoost
from xgboost import XGBClassifier
```

```python
# Import Evaluation Metrics
# - Accuracy	
# - Precision	
# - Recall 
# - F1-Score
# - Confusion Matrix	
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix, silhouette_score, davies_bouldin_score)
```

```python
logistic = LogisticRegression(penalty='l2', class_weight='balanced')
decisiontree = DecisionTreeClassifier(max_depth=5,  # Prevent overfitting
                                      min_samples_split=10,  # Minimum samples to split
                                      class_weight='balanced')
random = RandomForestClassifier(n_estimators=100, max_depth=10, min_samples_split=10,
                                class_weight='balanced_subsample'  # Handles imbalance
                               )
gradient = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3)
                                
adaboost = AdaBoostClassifier(n_estimators=50, learning_rate=1.0,)

gaussnb = GaussianNB()
multinb = MultinomialNB(alpha=1.0  # Smoothing parameter 
                       )
bernoulnb = BernoulliNB(alpha=1.0  # Smoothing parameter
                       )
xgb = XGBClassifier(n_estimators=100, max_depth=3,
                    learning_rate=0.1, subsample=0.8)
```

```python
clfs = {
  "LogisticRegression" : logistic,
  "DecisionTreeClassifier" : decisiontree,
  "RandomForestClassifier" : random,
  "GradientBoostingClassifier" : gradient,
  "AdaBoostClassifier" : adaboost,
  "GaussianNB" : gaussnb,
  "MultinomialNB" : multinb,
  "BernoulliNB" : bernoulnb,
  "XGBClassifier" : xgb
}
```

```python
def train_classifier(clf,X_train,y_train,X_test,y_test):
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    confusion = confusion_matrix(y_test, y_pred)

    return accuracy, precision, recall, f1, confusion
```

```python
# Run function for each model
for name, model in clfs.items():
    print(f"\nTraining {name}...\n")
    accuracy, precision, recall, f1, confusion = train_classifier(model, X_train, y_train, X_test, y_test)
    
    print(f"Results for {name}:")
    print(f"Accuracy Score is : {accuracy}")
    print(f"Precision Score is: {precision}")
    print(f"Recall Score is: {recall}")
    print(f"F1 Score is: {f1}")
    print(f"Confusion Matrix is: {confusion}")
    print("-" * 40)
```

```python
# Store results in a DataFrame
results = []

for name, model in clfs.items():
    accuracy, precision, recall, f1, confusion = train_classifier(model, X_train, y_train, X_test, y_test)
    results.append([accuracy, precision, recall, f1, confusion])
```

```python
# Create DataFrame
results_df = pd.DataFrame(results, columns=["Accuracy Score", "Precision Score", "Recall", "F1 Score", "Confusion Matrix"])
results_df
```
